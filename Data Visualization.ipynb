```python
#import data 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Horror = pd.read_csv('movie_mov_data_Horror.csv', header = None, names = ['movie_id','movie_title', 'Horror','user_id','rating','age','sex','occupation']) 
#讀取csv



#movei_item.info()
Horror
```

```python
Horror=Horror.drop(Horror.index[[0]]) #刪除第一行(第一行為文字)

Horror
```

```python
Horror = Horror.drop_duplicates(subset=["user_id"]) #刪除重複的使用者資料

Horror
```

```python
Horror = Horror.sort_values(["occupation"], ascending=1)
fig = plt.figure(figsize=(12,6))
plt.hist(Horror["occupation"],39) #看恐怖電影的職業分布
plt.xticks(rotation=-23) 
plt.title("Horror movie")
plt.show()
```

```python
fig = plt.figure(figsize=(2,5))
plt.hist(Horror["sex"],3) #看恐怖電影的性別分布
plt.title("Horror movie")
plt.show()
```

```python
Horror = Horror.sort_values(["age"], ascending=1)
fig = plt.figure(figsize=(15,7))
plt.hist(Horror["age"],95) #看恐怖電影的年齡分布
plt.title("Horror movie")
plt.show()
```
