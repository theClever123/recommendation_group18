```python
data_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
mov_data = pd.read_csv('u.data', sep='\t', names=data_cols ,encoding='latin-1')
mov_data.drop(['timestamp'], axis = 1, inplace = True)

mov_data.info()
```

```python
user_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
mov_user = pd.read_csv('u.user', sep='|', names=user_cols , encoding='latin-1', parse_dates=True) 
mov_user.drop(['zip_code'], axis = 1, inplace = True)

mov_user.info()
```

```python
m_cols = ['movie_id','movie_title','release date','video release date','IMDb URL', 
          'unknown','Action'   ,'Adventure','Animation','Childrens','Comedy' ,'Crime' ,'Documentary','Drama', 
          'Fantasy','Film-Noir','Horror'   ,'Musical'  ,'Mystery'  ,'Romance','Sci-Fi','Thriller'   ,'War'  ,'Western']
        
#m_cols = ['movie_id', 'movie_title', 'release date', 'video release date', 'IMDb URL','unknown']

mov_item = pd.read_csv('u.item', sep='|', names=m_cols, encoding = 'cp1252',header = None )

mov_item.drop(['release date'],       axis = 1, inplace = True)
mov_item.drop(['video release date'], axis = 1, inplace = True)
mov_item.drop(['IMDb URL'],           axis = 1, inplace = True)
mov_item.drop(['unknown'],            axis = 1, inplace = True)
mov_item.info()
```

```python
#movie_mov_data = pd.merge(mov_item, mov_data, left_index=True , right_index=True)
movie_mov_data = pd.merge(mov_item, mov_data)
movie_mov_data = pd.merge(movie_mov_data, mov_user)
movie_mov_data['sex']=movie_mov_data['sex'].map({'M':0 , 'F':1})
movie_mov_data.info()
#movie_mov_data.head(5)
#df.head(2)
```

```python
mov_data_matrix = movie_mov_data.pivot_table(index=['movie_id'],columns=['user_id'],values='rating').reset_index(drop=True)
mov_data_matrix.fillna( 0 , inplace = True )
mov_data_matrix.head()
```

```python
movie_similarity = 1 - pairwise_distances( mov_data_matrix.to_numpy(), metric="cosine") # ‘cosine’, ‘euclidean’...

np.fill_diagonal( movie_similarity, 0 ) #Filling diagonals with 0s for future use when sorting is done
mov_data_matrix = pd.DataFrame( movie_similarity )
mov_data_matrix.head(5)
```

```python
user_movie_name = "B. Monkey (1998)"
mov_id = mov_item[ mov_item['movie_title']==user_movie_name].index.tolist()
    
mov_id = mov_id[0]

mov_item['similarity'] = mov_data_matrix.iloc[mov_id]
#mov_item.columns = ['movie_id','movie_title','similarity']
mov_item.head(5)
mov_item.loc[ : ,  ['movie_id','movie_title','similarity'] ]
```

```python
a=mov_item.loc[ : ,  ['movie_id','movie_title','similarity'] ]
a=a.sort_values( ['similarity'], ascending = False)

print("Recommended movies ",mov_id ,": \n", a[1:6])


```
